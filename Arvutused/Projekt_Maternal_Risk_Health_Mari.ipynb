```python
#riskihinnangu numbrilise väärtuse lisamine andmestikku
risk_level_mapping = {'low risk': 1, 'mid risk': 2, 'high risk': 3}
data['RiskLevel_encoded'] = data['RiskLevel'].map(risk_level_mapping)
data.head()
```

```python
cleaned_data['RiskLevel_encoded'] = cleaned_data['RiskLevel'].map(risk_level_mapping)
cleaned_data.head()
```

```python
#teostame kruskal-wallise testi, et leida tunnuste mõju vastu riskihinnangut
from scipy.stats import kruskal

# Performing Kruskal-Wallis test for each column against 'RiskLevel'
kruskal_test_results = {}

# Splitting the data based on RiskLevel
groups = data.groupby('RiskLevel_encoded')

for column in data.columns:
    # Extracting the data for each group
    data_groups = [group[column].values for name, group in groups]
    statistic, p_value = kruskal(*data_groups)
    kruskal_test_results[column] = p_value

kruskal_test_results

results_df = pd.DataFrame(list(kruskal_test_results.items()), columns=['Variable', 'P-Value'])
results_df['Significant'] = results_df['P-Value'] <= 0.05

results_df
```

```python
#sama aga duplikaatideta andmestikule
# Performing Kruskal-Wallis test for each column against 'RiskLevel'
kruskal_test_results = {}

# Splitting the data based on RiskLevel
groups = cleaned_data.groupby('RiskLevel_encoded')

for column in cleaned_data.columns:
    # Extracting the data for each group
    data_groups = [group[column].values for name, group in groups]
    statistic, p_value = kruskal(*data_groups)
    kruskal_test_results[column] = p_value

kruskal_test_results
```

```python
#decisiontree duplikaatidega
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Splitting the data into features (X) and target (y)
X = data.drop(['RiskLevel', 'RiskLevel_encoded'], axis=1)
y = data['RiskLevel_encoded']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Training a Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_classifier.fit(X_train, y_train)

#Converting feature names to a list
feature_names_list = list(X.columns)

# Plotting the decision tree with the corrected feature names
plt.figure(figsize=(20,10))
plot_tree(dt_classifier, feature_names=feature_names_list, class_names=['Low Risk', 'Mid Risk', 'High Risk'], filled=True)
plt.show()
# Predicting on the test set
y_pred = dt_classifier.predict(X_test)

# Calculating accuracy and other performance metrics
accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)

accuracy, classification_report_output
```
